```python
data.sort_values("price", axis = 0, ascending = False,
                 inplace = True, na_position='last')
data
```

```python
toyota_cars = data.loc[data['company'] == 'toyota']
print(toyota_cars)
```

```python
jumlah_mobil_perusahaan = data.groupby('company').size().reset_index(name='jumlah_mobil')

# Tampilkan hasilnya dalam bentuk frame
print(jumlah_mobil_perusahaan.to_string(index=False))
```

```python
# Temukan mobil dengan harga tertinggi dari setiap perusahaan
mobil_tertinggi_perusahaan = data.loc[data.groupby('company')['price'].idxmax()]

# Tampilkan hasilnya
print(mobil_tertinggi_perusahaan)
```
