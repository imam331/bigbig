```python
import pandas as pd
```

```python
df = pd.read_csv('d:/data_soal_1.csv')

# Tampilan deskripsi statistik
print("Deskripsi Statistik:")
print(df.describe())

# Membuat grafik pencar
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Pendapatan'], df['Total'], alpha=0.5)
plt.title('Scatter Plot of Pendapatan vs Total')
plt.xlabel('Pendapatan')
plt.ylabel('Total')
plt.grid(True)
plt.show()

# Membuat grafik batang
plt.figure(figsize=(12, 6))
plt.bar(df['Nama'], df['Pendapatan'])
plt.title('Grafik Batang Pendapatan Berdasarkan Nama')
plt.xlabel('Nama')
plt.ylabel('Pendapatan')
plt.xticks(rotation=45)
plt.show()
```

```python
# Langkah a: Import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt

# Langkah b: Membuat dataframe
data = {
    'Id': [1, 2, 3, 4, 5, 6],
    'Year': [2017, 2018, 2019, 2020, 2021, 2022],
    'Bears': [8, 54, 93, 116, 137, 184],
    'Dolphins': [150, 77, 32, 11, 6, 1],
    'Whales': [80, 54, 100, 76, 93, 72]
}

df = pd.DataFrame(data)

# Langkah c: Tampilkan data teratas
print("Data teratas:")
print(df.head())

# Langkah d: Tampilkan data terbawah
print("\nData terbawah:")
print(df.tail())

# Langkah e: Tampilkan data statistik
print("\nData statistik:")
print(df.describe())

# Langkah f: Grafik garis untuk data Tahun dan Bear
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Bears'], marker='o', linestyle='-', color='b')
plt.title('Grafik Garis Bears per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Bears')
plt.grid(True)
plt.show()

# Langkah g: Grafik garis untuk Bears, Dolphins, dan Whales berdasarkan tahun
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Bears'], marker='o', linestyle='-', color='b', label='Bears')
plt.plot(df['Year'], df['Dolphins'], marker='o', linestyle='-', color='g', label='Dolphins')
plt.plot(df['Year'], df['Whales'], marker='o', linestyle='-', color='r', label='Whales')
plt.title('Grafik Garis Bears, Dolphins, dan Whales per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Populasi')
plt.legend()
plt.grid(True)
plt.show()

# Langkah h: Grafik area untuk data Whales
plt.figure(figsize=(10, 6))
plt.fill_between(df['Year'], df['Whales'], color='skyblue', alpha=0.4)
plt.plot(df['Year'], df['Whales'], color='Slateblue', alpha=0.6, linewidth=2)
plt.title('Grafik Area Whales per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Whales')
plt.grid(True)
plt.show()
```
